## L thuyt
Thut ton di truyn (Genetic Algorithm - GA) l mt phng php tm kim da trn c ch tin ha t nhin, ly cm hng t thuyt tin ha ca Darwin. GA c s dng  tm li gii gn ti u cho cc bi ton ti u ha phc tp.

### Cc bc chnh ca thut ton:
1. **Khi to qun th**: To ngu nhin mt tp hp cc c th (gii php tim nng).
2. **nh gi  thch nghi (Fitness Evaluation)**: Tnh gi tr hm mc tiu cho mi c th.
3. **La chn (Selection)**: Chn cc c th tt nht  lai ghp (thng s dng cc phng php nh Tournament Selection hoc Roulette Wheel).
4. **Lai ghp (Crossover)**: Kt hp hai c th  to ra c th mi.
5. **t bin (Mutation)**: Thay i ngu nhin mt s gen  duy tr tnh a dng.
6. **Lp li**: Quay li bc 2 cho n khi t iu kin dng (s th h hoc gi tr fitness mong mun).

```plaintext
To qun th mi rng
Lp cho n khi qun th mi  N c th:
    Chn cha m 1 bng Tournament Selection:
        - Chn ngu nhin tournament_size c th trong qun th
        - Chn c th c fitness cao nht lm cha m 1
    Chn cha m 2 bng Tournament Selection tng t
    
    Thc hin lai ghp (crossover) gia cha m 1 v cha m 2 vi xc sut crossover_rate:
        Nu crossover thnh cng:
            To con mi c gi tr trung bnh ca cha m

            Con mi l cha m 1
            
    Thc hin t bin cho con mi vi xc sut mutation_rate:

In ra c th tt nht v gi tr fitness tng ng ca th h


````python
import numpy as np
import random
import matplotlib.pyplot as plt

# Hm mc tiu
def fitness_function(x):
    return -(x**2) + 10*x + 50

# Khi to qun th
def initialize_population(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# La chn (Tournament Selection)
def select_parents(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghp
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

# t bin
def mutate(individual, mutation_rate=0.1, min_val=0, max_val=10):
    if random.random() < mutation_rate:
        return random.uniform(min_val, max_val)
    return individual

# Thut ton di truyn
def genetic_algorithm_example1(pop_size=50, generations=100, min_val=0, max_val=10):
    population = initialize_population(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function(x) for x in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents(population, fitness_values)
            parent2 = select_parents(population, fitness_values)
            child = crossover(parent1, parent2)
            child = mutate(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function(x) for x in population])]
        print(f'Th h {generation + 1}: x = {best_individual:.4f}, f(x) = {best_fitness:.4f}')
    
    best_fitness = max([fitness_function(x) for x in population])
    best_individual = population[np.argmax([fitness_function(x) for x in population])]
    
    # Trc quan ha
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Th h')
    plt.ylabel('Gi tr f(x) tt nht')
    plt.title('S thay i ca f(x) qua cc th h')
    plt.show()
    
    return best_individual, best_fitness

# Chy thut ton
best_x, best_f = genetic_algorithm_example1()
print(f'\nKt qu cui cng: x = {best_x:.4f}, f(x) = {best_f:.4f}')
````

# Bi tp 1
# Thay i hm mc tiu**: Th p dng thut ton cho hm $h(x) = sin(x) + cos(x)$ trong khong $x \in [0, 2\pi]$. So snh kt qu vi nghim chnh xc.

## Ti u hm $h(x) = \sin(x) + \cos(x)$ bng Thut Ton Di Truyn (Genetic Algorithm)

## Mc tiu
- Ti u hm:
  $$
  h(x) = \sin(x) + \cos(x)
  $$
  trong min:
  $$
  x \in [0, 2\pi]
  $$
- Tm gi tr **x** sao cho **h(x)** t gi tr ln nht.

---

## Phn tch nghim chnh xc

Ta c:
$$
h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xy ra khi:
$$
x = \frac{\pi}{4} \approx 0.7854
$$

---

##  tng thut ton di truyn (GA)

1. **M ha**: Biu din gi tr \( x \) bng chui nh phn di 16 bit.  
2. **Gii m**: Chuyn t chui nh phn sang gi tr thc trong khong \([0, 2\pi]\).  
3. **Fitness**: S dng chnh gi tr \( h(x) \) lm  thch nghi.  
4. **Chn lc**: Roulette Wheel Selection.  
5. **Lai ghp**: Lai 1 im vi xc sut cao.  
6. **t bin**: Lt bit vi xc sut thp.

---


````python
# code vo  y nhe
import numpy as np
import random
import matplotlib.pyplot as plt

# Hm mc tiu
def fitness_function(x):
    return np.sin(x) + np.cos(x)

# Khi to qun th
def initialize_population(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# La chn (Tournament Selection)
def select_parents(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghp (trung bnh 2 cha m)
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

# t bin
def mutate(individual, mutation_rate=0.05, min_val=0, max_val=2*np.pi):
    if random.random() < mutation_rate:
        return random.uniform(min_val, max_val)
    return individual

# Thut ton di truyn
def genetic_algorithm_h(pop_size=50, generations=100, min_val=0, max_val=2*np.pi):
    population = initialize_population(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function(x) for x in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents(population, fitness_values)
            parent2 = select_parents(population, fitness_values)
            child = crossover(parent1, parent2)
            child = mutate(child, mutation_rate=0.05, min_val=min_val, max_val=max_val)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function(x) for x in population])]
        print(f'Th h {generation + 1}: x = {best_individual:.4f}, h(x) = {fitness_function(best_individual):.4f}')
    
    best_fitness = max([fitness_function(x) for x in population])
    best_individual = population[np.argmax([fitness_function(x) for x in population])]
    
    # Trc quan ha
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Th h')
    plt.ylabel('Gi tr h(x) tt nht')
    plt.title('Tin ha h(x) qua cc th h')
    plt.grid()
    plt.show()
    
    return best_individual, best_fitness

# Chy thut ton
best_x, best_h = genetic_algorithm_h()
print(f'\n Kt qu cui cng: x = {best_x:.4f}, h(x) = {best_h:.4f}')
print(f' So vi nghim chnh xc: x = {np.pi/4:.4f}  0.7854, h(x) = {np.sqrt(2):.4f}  1.4142')
````

# Bi tp 2: Ti u hm $h(x) = \sin(x) + \cos(x)$ bng Thut Ton Di Truyn (Genetic Algorithm)

---

## Mc tiu

Ti u hm:
$$
h(x) = \sin(x) + \cos(x)
$$
trong khong:
$$
x \in [0, 2\pi]
$$

> Tm gi tr $x$ sao cho $h(x)$ t **gi tr ln nht**.

---

## Phn tch nghim chnh xc

Ta c:
$$
h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xy ra khi:
$$
x = \frac{\pi}{4} \approx 0.7854
$$

---

##   tng thut ton di truyn (GA)

| Thnh phn        | Chi tit                                                                 |
|------------------|-------------------------------------------------------------------------|
| **M ha**        | Dng chui nh phn 16 bit  biu din gi tr $x \in [0, 2\pi]$         |
| **Gii m**       | nh x nh phn sang gi tr thc: $x = 0 + \dfrac{\text{decimal}}{2^{16}-1} \cdot 2\pi$ |
| **Fitness**       | Chnh l $h(x) = \sin(x) + \cos(x)$                                     |
| **Chn lc**      | Roulette Wheel Selection                                                |
| **Lai ghp**      | 1 im ct vi xc sut 0.8                                              |
| **t bin**      | Xc sut 5% (0.05)                                                       |

---

````python
# code vo y nhe
import numpy as np
import matplotlib.pyplot as plt
import random

# Cu hnh thut ton
BIT_LENGTH = 16
POP_SIZE = 50
GENERATIONS = 100
CROSSOVER_RATE = 0.8
MUTATION_RATE = 0.05
MIN_X = 0
MAX_X = 2 * np.pi

# --------------------
# Hm chuyn i
# --------------------

# M ha: s thc  nh phn
def encode(x):
    max_int = 2**BIT_LENGTH - 1
    scaled = int((x - MIN_X) / (MAX_X - MIN_X) * max_int)
    return format(scaled, f'0{BIT_LENGTH}b')

# Gii m: nh phn  s thc
def decode(binary_str):
    max_int = 2**BIT_LENGTH - 1
    value = int(binary_str, 2)
    return MIN_X + (value / max_int) * (MAX_X - MIN_X)

# Hm fitness: h(x)
def fitness(x):
    return np.sin(x) + np.cos(x)

# Khi to qun th
def init_population():
    return [format(random.randint(0, 2**BIT_LENGTH - 1), f'0{BIT_LENGTH}b') for _ in range(POP_SIZE)]

# Chn lc: Roulette Wheel
def selection(population, fitnesses):
    total_fitness = sum(fitnesses)
    probs = [f / total_fitness for f in fitnesses]
    return random.choices(population, weights=probs, k=1)[0]

# Lai ghp 1 im
def crossover(p1, p2):
    if random.random() < CROSSOVER_RATE:
        point = random.randint(1, BIT_LENGTH - 1)
        return p1[:point] + p2[point:]
    return p1

# t bin
def mutate(binary_str):
    mutated = ''
    for bit in binary_str:
        if random.random() < MUTATION_RATE:
            mutated += '1' if bit == '0' else '0'
        else:
            mutated += bit
    return mutated

# GA chnh
def genetic_algorithm():
    population = init_population()
    best_fitness_history = []

    for gen in range(GENERATIONS):
        decoded = [decode(ind) for ind in population]
        fitnesses = [fitness(x) for x in decoded]

        best_fit = max(fitnesses)
        best_ind = population[np.argmax(fitnesses)]
        best_x = decode(best_ind)
        best_fitness_history.append(best_fit)

        print(f"Th h {gen+1:3d}: x = {best_x:.4f}, h(x) = {best_fit:.4f}")

        new_population = []
        for _ in range(POP_SIZE):
            parent1 = selection(population, fitnesses)
            parent2 = selection(population, fitnesses)
            child = crossover(parent1, parent2)
            child = mutate(child)
            new_population.append(child)

        population = new_population

    # Kt qu cui
    decoded = [decode(ind) for ind in population]
    fitnesses = [fitness(x) for x in decoded]
    best_index = np.argmax(fitnesses)
    best_x = decoded[best_index]
    best_h = fitnesses[best_index]

    # V  th
    plt.plot(best_fitness_history)
    plt.xlabel("Th h")
    plt.ylabel("Gi tr h(x) tt nht")
    plt.title("Tin ha h(x) qua cc th h")
    plt.grid()
    plt.show()

    return best_x, best_h

# Chy thut ton
best_x, best_h = genetic_algorithm()
print(f"\n Kt qu cui cng: x = {best_x:.4f}, h(x) = {best_h:.4f}")
print(f" So vi nghim chnh xc: x = {np.pi/4:.4f}  0.7854, h(x) = {np.sqrt(2):.4f}  1.4142")
````

# Thay i tham s thut ton di truyn v phn tch tc  hi t

**Mc tiu:** Th thay i cc tham s chnh ca thut ton di truyn gm:

- Kch thc qun th: `pop_size`
- S th h: `generations`
- T l t bin: `mutation_rate`
- T l lai ghp: `crossover_rate`

v phn tch nh hng ca tng tham s n tc  hi t ca thut ton.

---

## Cc tham s v vai tr

| Tham s         |  ngha                                  | nh hng chnh                                                                                         |
|-----------------|-----------------------------------------|--------------------------------------------------------------------------------------------------------|
| `pop_size`      | S c th trong qun th                 | Qun th ln gip khm ph khng gian tm kim tt hn nhng tn nhiu thi gian tnh ton. Qun th nh hi t nhanh nhng d b kt cc tr cc b. |
| `generations`   | S vng lp ca thut ton               | S th h cng nhiu th thut ton c nhiu c hi ci thin nghim, tuy nhin thi gian chy cng tng. |
| `mutation_rate` | Xc sut t bin mi bit                | t bin cao gip tng a dng gene, trnh b kt trong cc tr cc b, nhng nu qu cao lm qu trnh tm kim nhiu lon. |
| `crossover_rate`| Xc sut lai ghp gia hai c th       | Lai ghp gip phi hp gen tt, tng kh nng tm nghim tt. Nu qu thp, qun th km a dng; qu cao c th ph hy cu trc tt ca c th. |

---

## Phn tch nh hng

- **`pop_size`**: Kch thc qun th ln gip tm kim ton cc tt hn nhng chm hn, kch thc nh lm thut ton hi t nhanh nhng d mc kt.
- **`generations`**: S th h ln tng kh nng ci tin kt qu nhng lm tng thi gian tnh ton.
- **`mutation_rate`**: T l t bin va phi gip duy tr a dng di truyn, trnh kt cc b; qu cao hoc qu thp u lm gim hiu qu.
- **`crossover_rate`**: T l lai ghp hp l gip khai thc tt thng tin di truyn, tng hiu qu hi t.

---




````python
# code vo y 
import numpy as np
import random
import matplotlib.pyplot as plt

# Hm mc tiu
def fitness_function(x):
    return np.sin(x) + np.cos(x)

# Khi to qun th
def initialize_population(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# La chn (Tournament Selection)
def select_parents(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghp (trung bnh 2 cha m)
def crossover(parent1, parent2, crossover_rate):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

# t bin
def mutate(individual, mutation_rate, min_val, max_val):
    if random.random() < mutation_rate:
        return random.uniform(min_val, max_val)
    return individual

# Thut ton di truyn vi cc tham s ty chnh
def genetic_algorithm_custom(pop_size, generations, mutation_rate, crossover_rate, min_val=0, max_val=2*np.pi):
    population = initialize_population(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for _ in range(generations):
        fitness_values = [fitness_function(x) for x in population]
        best_fitness_history.append(max(fitness_values))
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents(population, fitness_values)
            parent2 = select_parents(population, fitness_values)
            child = crossover(parent1, parent2, crossover_rate)
            child = mutate(child, mutation_rate, min_val, max_val)
            new_population.append(child)
        
        population = new_population
    
    return best_fitness_history

# Chy th vi cc pop_size khc nhau  so snh tc  hi t
def run_experiment_popsize():
    pop_sizes = [10, 50, 100]
    results = {}

    for size in pop_sizes:
        print(f"ang chy vi pop_size = {size} ...")
        history = genetic_algorithm_custom(
            pop_size=size,
            generations=100,
            mutation_rate=0.05,
            crossover_rate=0.8
        )
        results[f"pop_size={size}"] = history

    # V biu  so snh
    plt.figure(figsize=(10, 6))
    for label, history in results.items():
        plt.plot(history, label=label)

    plt.xlabel("Th h")
    plt.ylabel("Gi tr h(x) tt nht")
    plt.title("So snh tc  hi t vi cc gi tr pop_size khc nhau")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    run_experiment_popsize()
````

##   tng ci tin

##  Kt qu v phn tch

- Phng php **Roulette Wheel** thng gip tng tc  hi t nh u tin c th tt.
- Tuy nhin c th lm gim a dng gen nu c th u t chim u th qu ln, d ri vo cc tr cc b.
- Phng php **Tournament** kim sot tt s a dng nhng c th chm hn trong mt s trng hp.
- Phng php **Random** n gin nhng t hiu qu trong ti u nng cao.

---


````python
# code vo y nh
import numpy as np
import random
import matplotlib.pyplot as plt

# Hm mc tiu
def fitness_function(x):
    return np.sin(x) + np.cos(x)

# Khi to qun th
def initialize_population(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# -------------------------
# PHNG PHP LA CHN
# -------------------------

# 1. Tournament Selection
def select_tournament(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# 2. Roulette Wheel Selection
def select_roulette(population, fitness_values):
    total_fitness = sum(fitness_values)
    if total_fitness == 0:
        return random.choice(population)  # fallback
    probs = [f / total_fitness for f in fitness_values]
    return random.choices(population, weights=probs, k=1)[0]

# Lai ghp (trung bnh)
def crossover(parent1, parent2, crossover_rate):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

# t bin
def mutate(individual, mutation_rate, min_val, max_val):
    if random.random() < mutation_rate:
        return random.uniform(min_val, max_val)
    return individual

# -------------------------
# THUT TON DI TRUYN
# -------------------------

def genetic_algorithm_selection(pop_size, generations, mutation_rate, crossover_rate, selection_method):
    min_val = 0
    max_val = 2 * np.pi
    population = initialize_population(pop_size, min_val, max_val)
    best_fitness_history = []

    for _ in range(generations):
        fitness_values = [fitness_function(x) for x in population]
        best_fitness_history.append(max(fitness_values))

        new_population = []
        for _ in range(pop_size):
            if selection_method == "tournament":
                parent1 = select_tournament(population, fitness_values)
                parent2 = select_tournament(population, fitness_values)
            elif selection_method == "roulette":
                parent1 = select_roulette(population, fitness_values)
                parent2 = select_roulette(population, fitness_values)
            else:
                raise ValueError("Unknown selection method")

            child = crossover(parent1, parent2, crossover_rate)
            child = mutate(child, mutation_rate, min_val, max_val)
            new_population.append(child)

        population = new_population

    return best_fitness_history

# -------------------------
# CHY TH & SO SNH
# -------------------------

def run_selection_experiment():
    methods = ["tournament", "roulette"]
    results = {}
    for method in methods:
        print(f"ang chy vi phng php chn: {method} ...")
        history = genetic_algorithm_selection(
            pop_size=50,
            generations=100,
            mutation_rate=0.05,
            crossover_rate=0.8,
            selection_method=method
        )
        results[method] = history

    # V biu  so snh
    plt.figure(figsize=(10, 6))
    for label, history in results.items():
        plt.plot(history, label=label)

    plt.xlabel("Th h")
    plt.ylabel("Gi tr h(x) tt nht")
    plt.title("So snh hiu qu la chn: Tournament vs Roulette")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    run_selection_experiment()
````

# Trc quan ha qun th  V biu  phn tn (scatter plot) th hin s phn b ca qun th qua cc th h

##  Mc tiu

- Trc quan ha s phn b ca qun th c th trong qu trnh tin ha ca thut ton di truyn.
- Dng biu  phn tn (scatter plot)  th hin v tr cc c th trn khng gian bin s qua tng th h.
- Gip quan st r tin trnh hi t, a dng di truyn, v s thay i qun th theo thi gian.

---

##  tng thc hin

- Vi bi ton ti u hm $k(x, y, z) = x^2 + y^2 + z^2$, khng gian tm kim l 3 chiu.
- Mi c th l mt im trong khng gian $(x, y, z)$.
-  mi th h, ly v tr gii m ca ton b c th, v scatter plot.
- Nu khng th trc quan 3D trc tip, c th v scatter 2D tng cp bin: $(x, y)$, $(y, z)$, $(x, z)$ qua cc th h.

---

````python
#code vo y
import numpy as np
import random
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Hm mc tiu
def fitness_function(x, y, z):
    return -(x**2 + y**2 + z**2)  # du tr  tm cc i

# Khi to qun th: mi c th l (x, y, z)
def initialize_population(pop_size, min_val, max_val):
    return [(
        random.uniform(min_val, max_val),
        random.uniform(min_val, max_val),
        random.uniform(min_val, max_val)
    ) for _ in range(pop_size)]

# La chn: Tournament
def select_tournament(population, fitnesses, k=3):
    selected = random.sample(range(len(population)), k)
    best = max(selected, key=lambda i: fitnesses[i])
    return population[best]

# Lai ghp trung bnh
def crossover(p1, p2, rate=0.8):
    if random.random() < rate:
        return tuple((a + b) / 2 for a, b in zip(p1, p2))
    return p1

# t bin
def mutate(ind, rate=0.05, min_val=-10, max_val=10):
    if random.random() < rate:
        return (
            random.uniform(min_val, max_val),
            random.uniform(min_val, max_val),
            random.uniform(min_val, max_val)
        )
    return ind

# Thut ton di truyn v lu qun th theo tng th h
def genetic_algorithm_3d(pop_size=50, generations=50, min_val=-10, max_val=10):
    population = initialize_population(pop_size, min_val, max_val)
    all_generations = []

    for gen in range(generations):
        fitnesses = [fitness_function(x, y, z) for x, y, z in population]
        all_generations.append(population.copy())

        new_population = []
        for _ in range(pop_size):
            p1 = select_tournament(population, fitnesses)
            p2 = select_tournament(population, fitnesses)
            child = crossover(p1, p2)
            child = mutate(child, min_val=min_val, max_val=max_val)
            new_population.append(child)

        population = new_population

    return all_generations

# V phn b qun th 3D
def visualize_generations_3d(all_generations):
    for gen_index, generation in enumerate(all_generations):
        fig = plt.figure(figsize=(7, 5))
        ax = fig.add_subplot(111, projection='3d')
        xs = [ind[0] for ind in generation]
        ys = [ind[1] for ind in generation]
        zs = [ind[2] for ind in generation]
        ax.scatter(xs, ys, zs, c='blue', alpha=0.6)
        ax.set_title(f'Phn b qun th - Th h {gen_index + 1}')
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        ax.set_zlabel('Z')
        plt.tight_layout()
        plt.show()

# Chy v hin th
if __name__ == "__main__":
    all_gens = genetic_algorithm_3d(pop_size=50, generations=10)
    visualize_generations_3d(all_gens)
````
